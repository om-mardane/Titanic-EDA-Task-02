```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].median())
```

```python
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.drop("Cabin", axis=1, inplace=True)
```

```python
df.isnull().sum()
```

```python
sns.countplot(x="Survived", data=df)
plt.title("Survival Count")
plt.show()
```

```python
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival by Gender")
plt.show()
```

```python
sns.countplot(x="Pclass", hue="Survived", data=df)
plt.title("Survival by Passenger Class")
plt.show()
```

```python
plt.hist(df["Age"], bins=30)
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
sns.boxplot(x="Survived", y="Fare", data=df)
plt.title("Fare vs Survival")
plt.show()
```

Observations:
1. Females had a significantly higher survival rate.
2. First-class passengers were more likely to survive.
3. Passengers who paid higher fares had better chances of survival.
4. Most passengers were aged between 2040 years.
5. Passenger class and gender strongly influenced survival.

Conclusion:
After performing data cleaning and exploratory data analysis,
it was observed that gender, passenger class, and fare were the
most influential factors affecting survival on the Titanic.
Proper data cleaning helped improve data quality and enabled
meaningful insights.
